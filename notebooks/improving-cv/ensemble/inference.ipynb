```python
import pandas as pd
import numpy as np
import kagglehub
import torch
import torch.nn as nn

from transformers import AutoTokenizer, AutoModel, AutoConfig
from torch.utils.data import Dataset, DataLoader
```

```python
class cfg:
    competition = 'learning-agency-lab-automated-essay-scoring-2'
    checkpoint = 'microsoft/deberta-v3-small''
    tokenizer = None
    max_length = 512
    batch_size = 16
    epochs = 10
    learning_rate = 5e-5
    weight_decay = 0.01
    task = 'regression'
```

```python
class EssayDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=cfg.max_length):
        self.X = data['full_text']
        if 'score' in data.columns:
            self.y = data['score']
        else:
            self.y = None
        self.tokenizer = tokenizer
        self.max_length = max_length
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        tokens = self.tokenizer(
            text = self.X.iloc[idx],
            max_length=self.max_length,
            truncation = True,
            padding = 'max_length',
            return_tensors = 'pt'
        )
        item = {
            'input_ids': tokens['input_ids'].squeeze(0),
            'attention_mask': tokens['attention_mask'].squeeze(0),
        }
        if self.y is not None:
            item['labels'] = torch.tensor(self.y.iloc[idx], dtype=torch.long)
        return item
```

```python
#MeanPooling code taken from online resources
class MeanPooling(nn.Module):
    def __init__(self):
        super(MeanPooling, self).__init__()

    def forward(self, last_hidden_state, attention_mask):
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
        sum_embeddings = torch.sum(last_hidden_state * input_mask_expanded, 1)
        sum_mask = input_mask_expanded.sum(1)
        sum_mask = torch.clamp(sum_mask, min=1e-9)
        mean_embeddings = sum_embeddings/sum_mask
        return mean_embeddings
```

```python
class EssayModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = AutoModel.from_pretrained(cfg.checkpoint)
        self.mp = MeanPooling()
        self.classifier = nn.Linear(self.model.config.hidden_size, 1)

    def forward(self, input_ids, attention_mask):
        output = self.model(input_ids = input_ids, attention_mask = attention_mask)
        # #CLS Pooling
        # cls_output = output.last_hidden_state[:, 0, :].float()
        # logits = self.classifier(cls_output)
        #Mean Pooling
        mp_output = self.mp(output.last_hidden_state, attention_mask)
        logits = self.classifier(mp_output)
        return logits
```

```python
#Get test.csv
path = kagglehub.competition_download(cfg.competition)
test_df = pd.read_csv(f'{path}/test.csv')

#Tokenizer
tokenizer = AutoTokenizer.from_pretrained(cfg.checkpoint)

#Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#Test
test_dataset = EssayDataset(test_df, tokenizer)
test_loader = DataLoader(test_dataset, batch_size=cfg.batch_size, shuffle = False)

all_preds = []

#Load both models
model_paths = ['/kaggle/input/notebooks/michelliez/base-deberta-essay-prediction-model/best_deberta_model_fold_0.pth', '/kaggle/input/notebooks/michelliez/essay-improvement-notebook-model/best_deberta_model_fold_0.pth']
for path in model_paths:
    deberta_model = EssayModel().to(device).float()
    deberta_model.load_state_dict(torch.load(path, map_location=device))

    #Inference
    deberta_model.eval()
        
    preds = []

    with torch.no_grad():
        for batch in test_loader:
            logits = deberta_model(batch['input_ids'].to(device), batch['attention_mask'].to(device)).squeeze(-1)
            preds.extend(logits.cpu().numpy())
            
    all_preds.append(np.array(preds))
    del deberta_model

raw_preds = np.mean(all_preds, axis=0)
thresholds = [0.84, 1.569, 2.596, 3.579, 4.364]
final_preds = np.digitize(raw_preds, thresholds)

#Add 1 back to scores
test_preds = [pred+1 for pred in final_preds]

#Submission
submission = pd.DataFrame({
    'essay_id': test_df['essay_id'],
    'score': test_preds
})
submission.to_csv('submission.csv', index=False)
```
